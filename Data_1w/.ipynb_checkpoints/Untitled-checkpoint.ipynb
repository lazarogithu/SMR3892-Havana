```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('d2.csv')
#IDs
ids_df = df['id'].unique().tolist()
ids_df
```

```python
def plot_ids(ids):
    
    df = pd.read_csv('d2.csv')

    plt.figure(figsize=(11, 10),dpi=300)

    colors = plt.cm.rainbow(np.linspace(0, 1, len(ids)))

    for i, id in enumerate(ids):
        df_id = df[df['id'] == id]
        plt.plot(df_id['x'], df_id['y'], 'o', color=colors[i], label=f'ID {id}')

    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.show()
```

```python
plot_ids(ids_df)
```
